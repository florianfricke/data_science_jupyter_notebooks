```python
import os, time
from os import listdir
from os.path import isfile, join
import os
import codecs
import re
```

```python
#Löschen von nicht benötigter Zeichen
def clean_text(text):
    text = re.sub(r'[$&{}@_*>\\#%+=\[\]]','', text)
    text = re.sub(r'[0-9][0-9][0-9][0-9]\n',' ', text)
    text = re.sub('\x0c','', text)
    text = re.sub('\ufeff','', text)   
    text = re.sub('\n',' ', text)
    text = re.sub('\r',' ', text)
    text = re.sub('a0','', text)
    text = re.sub('\'92t','\'t', text)
    text = re.sub('\'92s','\'s', text)
    text = re.sub('\'92m','\'m', text)
    text = re.sub('\'92ll','\'ll', text)
    text = re.sub('\'91','', text)
    text = re.sub('\'92','', text)
    text = re.sub('\'93','', text)
    text = re.sub('\'94','', text)
    text = re.sub('\.','. ', text)
    text = re.sub(' +',' ', text)
    return text
```

```python
path = "../../Daten/txt/"
book_files = [f for f in listdir(path) if isfile(join(path, f))]

def get_books_content(batch_size, book_files, path):
    part_of_book = []
    for i in range(len(book_files)):
        input_file = path + book_files[i]
        file_position = 0
        f = open(input_file, 'r', encoding="utf8")
        filesize = os.fstat(f.fileno()).st_size # Anzahl der Zeichen in der Datei
        while(True):
            if ((file_position + batch_size) > filesize):
                break
                f.close()
            else:
                part_of_book.append(clean_text(read_file(f, file_position, batch_size)))
                file_position = f.tell()
    return part_of_book

def read_file (f, start_position, number_char):
    f.seek(start_position)
    return f.read(number_char)

part_of_book = get_books_content(100, book_files, path)
```
