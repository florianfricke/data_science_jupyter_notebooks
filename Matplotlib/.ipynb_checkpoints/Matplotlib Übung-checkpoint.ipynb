## Matplotlib Ãœbung

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,z)
axes2 = fig.add_axes([0.2,0.5,0.2,0.2])
axes2.plot(x,y)
axes2.set_xlim([0,22])
axes2.set_ylim([30,50])
```

```python
dia, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y,'b--', linewidth=5)
axes[1].plot(x,z, 'r',linewidth=4)
plt.tight_layout()
```
