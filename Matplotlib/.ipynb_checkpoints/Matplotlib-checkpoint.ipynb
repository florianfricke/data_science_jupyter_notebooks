```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#notwendig um im Notebook die Plots zu sehen
#falls mein keine Notebook verwendet sondern ein Editor gibt man am Ende plt.show() ein --> Graph öffnet sich in einem neuen Fenster
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y,'#32e5bb') #kann auch statt hexwert Buchstaben eingeben z.B. r = Rot
plt.xlabel('x-Achse')
plt.ylabel('y-Achse')
plt.title('Diagrammtitel')
#plt.show() -> falls kein Notebook verwendet wird
```

### Subplots: mehrere Diagramme nebeneinander ausgeben

```python
plt.subplot(1,2,1) #Anzahl der Zeilen, Anz. Spalten, Plotnummer -> 1.Plot
plt.plot(x,y,'b-*') #Linienart
plt.subplot(1,2,2) #2. Plot
plt.plot(y,x,'g.')
```

### objektorientierte Methode 
können Lage und Größe des Plots bestimmen

**axes = subplot**

```python
af = plt.figure() #af = leere Arbeitsfläche erstellen
axes = af.add_axes([0.1,0.1,0.4,0.8]) #Pos. links unten, Pos. unten, Breite der af, Höhe der af
axes.plot(x,y)
axes.set_xlabel('X Achse')
axes.set_ylabel('Y Achse')
axes.set_title('Überschrift')
```

```python
af = plt.figure()
axes1 = af.add_axes([0.1,0.1,0.8,0.8])
axes2 = af.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_title('großes Plot')
axes2.plot(y,x,'g')
axes2.set_title('kleines Plot')
```

```python
diag, axes = plt.subplots(nrows = 1, ncols = 2) #es soll 2 Diagramme in einer Zeile und innerhalb 2 Spalten geben
for ax in axes:
    ax.plot(x,y)
    ax.set_xlabel('X Achse')
    ax.set_ylabel('Y Achse')
    
plt.tight_layout() #verhindert das es keine Überlappungen gibt
```

```python
diag, axes = plt.subplots(nrows = 1, ncols = 2) #es soll 2 Diagramme in einer Zeile und innerhalb 2 Spalten geben
axes[0].plot(x,y,'b')
axes[0].set_title('1. Plot')
axes[1].plot(y,x,'g')
axes[1].set_title('2. Plot')
```

### Diagrammgröße

```python
diag = plt.figure(figsize=(3,0.8), dpi=150) #8 = Breite, 4 = Höhe, dpi = Pixel pro Zoll 
ax = diag.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#diag, axes = plt.subplots(nrows = 1, ncols = 2, figsize=(3,0.8),dpi=150) --> auch verwendbar für subplots
```

### Texte & Annotationen

```python
plt.arrow(0,0,0.5,0.5)
plt.text(1.5,1, 'Text der eingefügt wurde', style='italic')
plt.annotate('hier ist die Kurve zu Ende', xy=(0.5, 0.5), xytext=(0.7, 0.7),arrowprops=dict(facecolor='black', shrink=0.05),)
```

### Diagramm als Datein abspeichern

```python
af.savefig('Diagramm.png', dpi = 200) #Speichern des Plots
```

### Legende

```python
diag = plt.figure()
ax = diag.add_axes([0,0,0.5,0.5])
ax.plot(x, x**2, label = "x**2")
ax.plot(x, x**3, label = "x**3")
ax.legend(loc=6) #Position der Legende bestimmen mit loc --> Positionen stehen in der Api
```

### Kurven festlegen

```python
diag, ax = plt.subplots()
ax.plot(x, x**2, color="blue", alpha=0.9, linewidth=3, linestyle="steps", 
        marker='s', markersize=6, markerfacecolor='r', markeredgewidth=2, markeredgecolor='g') 
#alpha = transparenz, linewidth oder lw = Linienbreite, linestyle oder ls, marker (o,s,+), markeredgewidth = Rand der Marker
```

### Diagrammbereich / Diagrammauschnitte

```python
diag, ax = plt.subplots()
ax.plot(x, x**2, color="blue")
ax.set_ylim([0,5]) #y-Achse von Wert 0-5
ax.set_xlim([0,2]) #y-Achse von Wert 0-5
```
