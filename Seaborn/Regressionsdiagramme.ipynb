## Regression Plots
*lmplot()* ermöglicht es lineare Modelle zu visualisieren. Zusätzlich werden aber auch Funktionen zum Aufteilen und Einfärben von Daten anhand einzelner Variablen gegeben.

```python
import seaborn as sns
import pandas as pd
%matplotlib inline
tips = pd.read_csv('../Daten/csv/tips.csv')
tips.head(2)
```

### lmplot()

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="smoker")
```

### regplot() -> Markierungen setzen
= allgemeinere Form des lmplot(), regplot hat scatter_kws Parameter --> innerhalb dessen definieren wir s in einem Dictionary, um die Markierungsgröße festzulegen --> http://matplotlib.org/api/markers_api.html

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','v'], scatter_kws={'s':80}) 
#o = Kreis, v = Dreieck | scatter_kws = Größe der Pkt./Markierungen
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="smoker") 
#mit col statt hue werden die Daten nicht in einem sondern in mehreren Dia angezeigt
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="smoker", row="time") 
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="smoker", row="time", hue="day") #im Dia. werden noch die Tage angezeigt
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="day", hue="smoker", aspect=0.6)
```
