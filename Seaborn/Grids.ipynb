```python
grid = sns.PairGrid(tips)
#Platzierung der Plots √§ndern
grid.map_diag(plt.hist) #auf der Diagonalen sind Histogramme
grid.map_upper(plt.scatter) #alle Dia √ºber der Diagonalen
grid.map_lower(sns.kdeplot) #alle Dia unter der Diagonalen
```

### FacetGrid
= allgemeine Weg um eine Vielzahl an Diagrammen in einem Grid zu erstellen

```python
grid = sns.FacetGrid(tips, col="time", row="smoker", hue="sex")
#grid.map(plt.hist, 'total_bill') #hist ben√∂tigt 1 Argument
grid.map(plt.scatter, 'total_bill', 'tip').add_legend() #scatter ben√∂tigt 2 Argumente
```

### JointGrid
JointGrid ist die allg. Version f√ºr jointplot()

```python
grid = sns.JointGrid(x="total_bill", y="tip", data=tips)
grid.plot(sns.regplot, sns.distplot) #in einem Grid werden 2 verschiedene Dia. / Plots dargestellt
```
