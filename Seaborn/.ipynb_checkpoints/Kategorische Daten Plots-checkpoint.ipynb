## Kategorische/Nominale Daten Plots

```python
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
tips = pd.read_csv('../Daten/csv/tips.csv')
tips.head()
```

### Barplot
Nominale/ Kategorische Daten (z.B. day) können mit ihrem Durschnitt angezeigt werden

```python
sns.barplot(x="day", y="total_bill", data=tips) #Durschnittsrechnungsbetrag für Samstag ist 20
```

```python
sns.barplot(x="day", y="total_bill", data=tips, estimator=np.std) 
#eigene Berechnungen verwenden --> nicht mehr Durschnitt sondern Standartabweichung wird angezeigt
```

### countplot
= gleiche Dia wie barplot, jedoch zählt countplot die Anzahl (count) an Erscheinungen --> deshalb nur einen x Wert übergeben

```python
ax = sns.countplot(x='day', data=tips) #Samstag
ax.set(ylim=(0, len(tips[tips['day']=="Sat"])+2))
```

### boxplot
zeigt Verteilung für Kategorische/Nominale Daten --> Quantile, Ausreißer (engl. Outlier)

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="rainbow", orient="v") #orient = h --> horizontale Anordnung
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker", palette="Set1")
```

### violinplot
zeigt Verteilung von Daten entlang mehrerer Werte einer kategorischen/nominalen Variablen
ähnlich boxplot -> Unterschied: box plot repräsentiert tatsächliche Datenpunkte, violin plot eine KDE der zugrundeliegenden Verteilung

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="smoker", split=True) #mit split werden die Merkmale von Smoker zusammen gelegt
```

### stripplot
= Scatterplot für kategorische/nominale Variablen

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue="smoker") #mit Jitter werden die Punkte auseinander gezogen
```

### swarmplot
ähnlich stripplot -> Unterschied: Punkte überschneiden sich nicht

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="smoker")
```

### Kategorische/Nominale Plots kombinieren

```python
sns.violinplot(x="tip", y="day", data=tips)
sns.swarmplot(x="tip", y="day", data=tips, color="black", size=2)
```

### factorplot
mit kind Parameter kann man die  Diagrammart aussuchen

```python
sns.factorplot(x="smoker", y="total_bill", data=tips, kind="bar")
```
