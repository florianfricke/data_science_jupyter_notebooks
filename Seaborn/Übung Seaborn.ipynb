```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
%matplotlib inline
t = pd.read_csv('../Daten/csv/titanic.csv')
t.head(3)
```

```python
sns.jointplot(x="fare", y="age", data=t, kind="scatter")
```

```python
sns.boxplot(x="class", y="age", data=t)
```

```python
sns.swarmplot(x='class', y='age', data=t)
```

```python
sns.countplot(x='sex', data=t)
```

```python
sns.heatmap(t.corr(), cmap='coolwarm',annot=True)
```

```python
sns.distplot(t["fare"], kde=False)
```
